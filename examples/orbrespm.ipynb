```python
import numpy as np
import matplotlib.pyplot as plt
import pyacal
from pyacal.experiments.orbrespm import OrbRespm
```

```python
pyacal.set_facility('sirius')
```

```python
orbrespm = OrbRespm(accelerator='SI', isonline=True)
```

```python
orbrespm.params.delta_freq = 100   # [Hz]
orbrespm.params.delta_curr_hcm = 0.1  # [A]
orbrespm.params.delta_curr_vcm = 0.1  # [A]
orbrespm.params.sofb_nrpoints = 20
print(orbrespm.params)
print(f"Connected? {orbrespm.connected}")
```

```python
sofb = orbrespm.devices["sofb"]

# Select which CMs to measure
sofb.hcm_enbl[6:] = False
sofb.vcm_enbl[8:] = False
```

```python
orbrespm.start()
```
