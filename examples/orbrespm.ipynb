# Example on how to run Orbit response measurement.

Code specific for SIRIUS is commented.

```python
import numpy as np
import matplotlib.pyplot as plt
import pyacal
from pyacal.experiments.orbrespm import OrbRespm
```

```python
pyacal.set_facility('sirius')
```

```python
orbrespm = OrbRespm(accelerator='SI', isonline=True)
```

```python
orbrespm.params.delta_freq = 100   # use value with appropriated units here
orbrespm.params.delta_stren_hcm = 0.1 # use value with appropriated units here
orbrespm.params.delta_stren_vcm = 0.1 # use value with appropriated units here
orbrespm.params.sofb_nrpoints = 20
print(orbrespm.params)
print(f"Connected? {orbrespm.connected}")
```

```python
sofb = orbrespm.devices["sofb"]

# Select which CMs to measure
sofb.hcm_enbl[6:] = False
sofb.vcm_enbl[8:] = False
sofb.rfg_enbl = True
```

```python
orbrespm.start()
```
